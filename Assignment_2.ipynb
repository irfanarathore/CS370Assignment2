```python
import os
import torch

# Define training parameters
max_epochs = 18
save_every_n_epochs = 5
checkpoint_dir = "./checkpoints"
os.makedirs(checkpoint_dir, exist_ok=True)

# Initialize Engine and models for each category
engine = Engine(max_epochs = 18)
models = {}
start_epochs = {}

# Load models and determine start epochs
for category in categories:
    models[category] = EfficientAd()
    checkpoint_path = os.path.join(checkpoint_dir, f"efficientad_{category}.pth")

    if os.path.exists(checkpoint_path):
        print(f"Loading checkpoint for {category}...")
        checkpoint = torch.load(checkpoint_path, map_location="cuda" if torch.cuda.is_available() else "cpu")
        models[category].load_state_dict(checkpoint["model_state_dict"])
        start_epochs[category] = checkpoint.get("epoch", 0) + 1  # Resume from last saved epoch
    else:
        start_epochs[category] = 0  # Start from scratch if no checkpoint exists

# Train all categories together
for epoch in range(max(start_epochs.values()), max_epochs):  # Start from latest saved epoch
    print(f"Epoch {epoch+1}/{max_epochs} - Training all categories")

    for category in categories:
        if epoch < start_epochs[category]:
            continue  # Skip already trained epochs

        print(f"Training on {category} (Epoch {epoch+1})")
        datamodule = MVTec(category=category, num_workers=0, train_batch_size=1)
        model = models[category]

        engine.fit(model=model, datamodule=datamodule)

        # Save checkpoint every n epochs
        if (epoch + 1) % save_every_n_epochs == 0:
            checkpoint_path = os.path.join(checkpoint_dir, f"efficientad_{category}.pth")
            print(f"Saving checkpoint for {category} at epoch {epoch+1}...")
            torch.save({
                "epoch": epoch,
                "model_state_dict": model.state_dict(),
            }, checkpoint_path)

# Evaluate after training is complete
for category in categories:
    print(f"Evaluating EfficientAD on {category}")
    test_results = engine.test(
        model=models[category],
        datamodule=MVTec(category=category, num_workers=0, train_batch_size=1),
        ckpt_path=engine.trainer.checkpoint_callback.best_model_path,
    )
    print(f"Results for {category}: {test_results}")
```

```python
import torchvision.models as models

# Load a pretrained feature extractor
feature_extractor = models.resnet50(pretrained=True)
feature_extractor = torch.nn.Sequential(*list(feature_extractor.children())[:-1])  # Remove classification head
feature_extractor.eval()
import torchvision.transforms as transforms
from PIL import Image
import torch

# Define preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def extract_features(image_path, model):
    """Extract features from an image using the model."""
    img = Image.open(image_path).convert("RGB")
    img = transform(img).unsqueeze(0)  # Add batch dimension

    with torch.no_grad():
        features = model(img)

    return features.squeeze().numpy()  # Convert to NumPy array
```

```python
from qdrant_client.models import Distance, VectorParams, PointStruct

# Connect to Qdrant (local or cloud)
client = QdrantClient(":memory:")  # Use ":memory:" for an in-memory DB or provide a real endpoint

# Create a collection
client.recreate_collection(
    collection_name="anomaly_features",
    vectors_config=VectorParams(size=2048, distance=Distance.COSINE),  # Adjust size based on model output
)
```

```python
def add_to_qdrant(image_path, image_id, model):
    """Extract features from an image and store in Qdrant."""
    vector = extract_features(image_path, model)

    client.upsert(
        collection_name="anomaly_features",
        points=[PointStruct(id=image_id, vector=vector)]
    )

# Example: Add multiple images
image_files = ["/content/datasets/MVTec/grid/test/bent/000.png", "/content/datasets/MVTec/grid/test/broken/000.png", "/content/datasets/MVTec/grid/test/good/000.png"]  # Replace with actual image paths
for i, img_path in enumerate(image_files):
    add_to_qdrant(img_path, image_id=i, model=feature_extractor)
```

```python
def search_similar_images(query_image_path, model, top_k=5):
    """Find the top-k most similar images in Qdrant."""
    query_vector = extract_features(query_image_path, model)

    search_result = client.search(
        collection_name="anomaly_features",
        query_vector=query_vector,
        limit=top_k
    )

    # Return the IDs of similar images
    return [hit.id for hit in search_result]

# Example usage
similar_images = search_similar_images("/content/datasets/MVTec/grid/test/bent/000.png", model=feature_extractor, top_k=5)
print("Top 5 similar images:", similar_images)
```
